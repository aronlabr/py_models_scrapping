```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer['feature_names']
```

```python
cancer['target_names']
```

```python
cancer_df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
cancer_df.head()
```

```python
df_target = pd.DataFrame(cancer['target'],columns=['Cancer'])
```

```python
df = cancer_df.join(df_target)
```

```python
df.head()
```

```python
df['Cancer'].value_counts()
```

```python
#Estandarización
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('Cancer',axis=1))
```

```python
scaled_features = scaler.transform(df.drop('Cancer',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

```python
df_res = df[['Cancer']]
df_res.head()
```

```python
df_escala = pd.concat([df_feat,df_res],axis=1)
```

```python
df_escala.head()
```

```python
#División del conjunto, entrenamiento y prueba
from sklearn.model_selection import train_test_split
```

```python
X_escala = df_escala.drop('Cancer',axis=1)
y = df_escala['Cancer']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_escala,
                                                    y,
                                                    test_size=0.30,
                                                    random_state=50)
```

```python
#Aplicando KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
#Predicciones y evaluaciones
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
#Elaboracion del K
pred=[0,1,0,1,0]
real=[0,1,0,1,1]
c=np.mean(np.array(pred) != np.array(real))
print(c)
```

```python
tasa_error = []

for i in range(1,61):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    tasa_error.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,61),tasa_error,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Tasa de Error vs. Valor de K')
plt.xlabel('K')
plt.ylabel('Tasa de Error')
```

Aquí podemos ver que después de aplicar K> 20, la tasa de error tiende a rondar los 0.035. Reentrenemos el modelo con este dato y verifiquemos el informe de clasificación.

```python
# PRIMERO UNA COMPARACIÓN RÁPIDA CON EL K=1 ORIGINAL
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('CON K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
# AHORA CON K=20
knn = KNeighborsClassifier(n_neighbors=20)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('CON K=20')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
